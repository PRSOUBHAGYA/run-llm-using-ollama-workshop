# <font color="#007BFF">**Introduction to Ollama** </font>

**Ollama** is an open-source framework designed to run large language models (LLMs) locally. It streamlines the process of downloading, managing, and interacting with models through a dedicated Python library and a local REST API.

## <font color='#008080'>**1. Importing Library** </font>

To interact with the Ollama service programmatically, we use the official `ollama` Python SDK. This library handles the communication between your script and the local Ollama server running on your machine.

To convert Images to base64 encoded string we will use `base64` Python Package.

```python
import ollama
import base64
```

## <font color='#008080'>**2. Generating Response**

The `generate` method is the core function for obtaining text completions. It is a **stateless** request where you provide a model name and a prompt. The model processes the input and returns a structured response containing the generated text and execution metadata.

```python
response = ollama.generate(model='llama3.2:latest', prompt='Explain LLM to a 10th Class student in 5 lines.')
```

```python
response
```

```python
print(response.response)
```

## <font color='#008080'>**3. Streaming the Response**

By enabling the `stream` parameter, the model returns an **iterable generator**. This allows the application to receive and display tokens in real-time as they are produced, rather than waiting for the entire block of text to be finalized, which significantly improves the user experience.

```python
response = ollama.generate(model='llama3.2:latest', prompt='Explain LLM to a 10th Class student',stream=True)
for res in response:
    print(res.response,end='')
```

## <font color='#008080'>**4. Working with Images**

Ollama supports **Multimodal LLMs** (such as *gemma*,*Llava*), which are capable of processing both text and visual data simultaneously to perform "Visual Question Answering" (VQA).

### <font color='#008080'>**4.1. Single Image**

```python
image_path = 'transformer_architecture.jpg'
```

```python
with open(image_path,"rb") as f: # operning image in binary mode 
    image_bytes= f.read()    # reading the image 
image_64= base64.b64encode(image_bytes).decode("utf-8") # encoding the image in 64 bits.
```

```python
response = ollama.generate(model='gemma3:4b', prompt='Explain the image briefly',images=[image_64],stream=True)
for res in response:
    print(res.response,end='')
```

### <font color='#008080'>**4.2. Multiple Images**

```python
image_paths = ['transformer_architecture.jpg','gemma_benchmark_score.jpg']
base_64_images = []
for each_image_path in image_paths:
    with open(each_image_path,"rb") as f: 
        image_bytes= f.read()
    image_64= base64.b64encode(image_bytes).decode("utf-8")
    base_64_images.append(image_64)
```

```python
response = ollama.generate(model='gemma3:4b', prompt='Explain the images briefly',images=base_64_images,stream=True)
for res in response:
    print(res.response,end='')
```

## <font color='#008080'>**5. Hyperparameter Tuning**

Model behavior can be customized using the `options` dictionary. These parameters allow you to fine-tune the "creativity," deterministic nature, and resource usage of the local LLM.

* **`temperature`**: Adjusts the randomness of the output. Higher values (e.g., $1.5$) make the model more creative, while lower values (e.g., $0.1$) make it more deterministic. (Default: $0.8$)
* **`top_p`**: Nucleus sampling; limits the model to consider tokens whose cumulative probability reaches $P$. (Default: $0.9$)
* **`top_k`**: Limits the vocabulary to the $K$ most likely next words, reducing the chance of nonsense. (Default: $40$)
* **`min_p`**: Sets a minimum probability threshold relative to the most likely token to balance quality and variety.
* **`repeat_penalty`**: Penalizes repetitions. Higher values (e.g., $1.2$) discourage the model from repeating phrases. (Default: $1.1$)
* **`repeat_last_n`**: Sets how far back the model looks to prevent repetition. $0$ disables it; $-1$ looks at the entire context. (Default: $64$)
* **`presence_penalty`**: Encourages the model to talk about new topics by penalizing tokens that have already appeared.
* **`frequency_penalty`**: Specifically targets repetitive word usage based on how often they have appeared.
* **`num_ctx`**: Sets the size of the context window (in tokens). This is the model's "memory" for the current conversation. (Default: $2048$)
* **`num_predict`**: The maximum number of tokens to generate. Set to $-1$ for infinite generation until a stop token is hit.
* **`stop`**: A list of specific strings that will trigger the model to stop generating immediately.
* **`seed`**: Sets the random number seed. Using a specific integer (e.g., $42$) makes the output reproducible.
* **`num_thread`**: Sets the number of CPU threads used. Ollama usually optimizes this automatically.
* **`mirostat`**: Enables Mirostat sampling to control perplexity ($0$ = disabled, $1$ = Mirostat, $2$ = Mirostat 2.0).

```python
response = ollama.generate(model='llama3.2:latest', prompt='Explain LLM to a 10th Class student',stream=True, options={
    'temperature':0.9,
    'top_p':0.1
})
for res in response:
    print(res.response,end='')
```

## <font color='#008080'>**6. Basic Ollama Commands**

The library provides several management functions that mirror the CLI commands, allowing for automated model administration within your Python environment.

### <font color='#008080'>**6.1. Available LLMs in Your Machine**

The `list` function retrieves a catalog of all models currently stored on your local disk. It includes details such as the model version, file size, and the date it was last modified.

```python
local_models= ollama.list()
print(local_models)
print("**************************")

for i in local_models["models"]:
    print(i["model"], end='   ')
    print(str(round(i["size"]/1073741824,2))+"GB")
```

### <font color='#008080'>**6.2. Download LLM from Ollama Hub**

The `pull` function is used to download new model weights from the **Ollama Registry**. It fetches the required layers and configures them for your specific hardware (CPU/GPU) automatically.

```python
model_name="deepseek-r1"
progess = ollama.pull(model_name, stream=True)

for i in progess:
    print(i)
```

### <font color='#008080'>**6.3. Details of an LLM**

The `show` function provides a deep dive into a specific model's **Modelfile**. This includes the underlying architecture, the license information, and the specific **System Prompt** used to define the AI's persona.

```python
models_details = ollama.show("gemma3:4b")
print(models_details.dict())
```

```python
model_dict= models_details.dict()
print(model_dict["capabilities"])
print(model_dict["parameters"])
```
